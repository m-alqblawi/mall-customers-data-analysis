```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
mall_customers = pd.read_csv('./dataset/Mall_Customers.csv')
```

# 1. Customer Age Distribution Analysis: 

### What is the distribution of customer ages?

```python
# Age Distribution
sns.histplot(mall_customers['Age'], bins=20, kde=True)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# 2. Gender-Based Analysis:

### How does spending score vary between male and female customers?

```python
# Spending Score by Gender
sns.boxplot(x='Gender', y='Spending Score (1-100)', data=mall_customers)
plt.title('Spending Score by Gender')
plt.show()
```

# 3. Income and Spending Relationship:

### Is there a correlation between annual income and spending score?

```python
# Income and Spending Score Relationship
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=mall_customers, hue='Gender')
plt.title('Income vs. Spending Score')
plt.show()
```

# 4. Customer Segmentation:

### Can we identify distinct customer segments based on age, income, and spending score?

```python
from sklearn.cluster import KMeans

# Select relevant features for segmentation
features_for_segmentation = mall_customers[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Use k-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
mall_customers['Segment'] = kmeans.fit_predict(features_for_segmentation)

# Visualize the segments
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=mall_customers, hue='Segment')
plt.title('Customer Segmentation')
plt.show();
```

# 5. Income Distribution:

### What is the distribution of annual income among customers?

```python
# Annual Income Distribution
sns.histplot(mall_customers['Annual Income (k$)'], bins=20, kde=True)
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Frequency')
plt.show()
```

# 6. Age and Spending Score Relationship:

### How does spending score vary with age?

```python
# Age and Spending Score Relationship
sns.scatterplot(x='Age', y='Spending Score (1-100)', data=mall_customers, hue='Gender')
plt.title('Age vs. Spending Score')
plt.show()
```

# 7. Annual Income and Age Distribution by Gender:

### How is annual income distributed across different age groups, and is there a difference between genders?

```python
# Annual Income and Age Distribution by Gender
sns.scatterplot(x='Age', y='Annual Income (k$)', data=mall_customers, hue='Gender')
plt.title('Annual Income and Age Distribution by Gender')
plt.show()
```

# 8. Customer Spending Score Distribution:

### What is the distribution of spending scores among customers?

```python
# Spending Score Distribution
sns.histplot(mall_customers['Spending Score (1-100)'], bins=20, kde=True)
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Frequency')
plt.show()
```

# 9. Age and Income Distribution by Customer Segment:

### How are age and annual income distributed among different customer segments?

```python
# Age and Income Distribution by Segment
sns.scatterplot(x='Age', y='Annual Income (k$)', data=mall_customers, hue='Segment')
plt.title('Age and Income Distribution by Segment')
plt.show()
```

# 10. Customer Spending Score Analysis by Segment:

### What is the average spending score for each customer segment?

```python
# Spending Score Analysis by Segment
segment_avg_spending = mall_customers.groupby('Segment')['Spending Score (1-100)'].mean().reset_index()
sns.barplot(x='Segment', y='Spending Score (1-100)', data=segment_avg_spending)
plt.title('Average Spending Score by Segment')
plt.show()
```
